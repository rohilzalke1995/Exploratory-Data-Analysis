### Handling Missing Values

```python
data['Memory'].fillna(data['Memory'].mode()[0], inplace=True)
data['Storage'].fillna(data['Storage'].mode()[0], inplace=True)
data['Color'].fillna(data['Color'].mode()[0], inplace=True)
```

```python
MODE = data[data['Brand']== "Google Pixel"]['Model'].mode()
```

```python
data['Model'].fillna(data[data['Brand']== "Google Pixel"]['Model'].mode()[0], inplace=True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data['Original Price'] = data.groupby(['Brand', 'Model'])['Original Price'].transform(lambda x: x.fillna(data.loc[x.index, 'Selling Price'].median()))
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['Selling Price'] = data['Selling Price'].fillna(data['Selling Price'].median())
```

### Convert Data Type

```python
data['Memory'] = pd.to_numeric(data['Memory'].str.extract(r'(\d+)')[0], errors = 'coerce')
data['Storage'] = pd.to_numeric(data['Storage'].str.extract(r'(\d+)')[0], errors = 'coerce')
```

```python
data.isnull().sum()
```

```python
data['Memory'] = data['Memory'].astype('int')
data['Storage'] = data['Storage'].astype('int')
```

### 1. What are the different price range segments for mobiles in India?

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.figure(figsize=(10,10))
sns.histplot(data['Selling Price'], bins=30, kde=True)
plt.title('Distribution of Selling Price')
plt.xlabel('Selling Price in Rs.')
plt.ylabel('Number of Mobiles')
plt.xticks(np.arange(0, data['Selling Price'].max()+1, 5000), rotation=90)  # Change 2000 to 1000 or 500 for finer ticks
plt.yticks(np.arange(0, plt.gca().get_ylim()[1]+1, 50))        # Change 10 to 5 or 2 for finer ticks
plt.grid(True)

plt.show()
```

```python
def Price_Segment(Price):
    if Price<= 10000:
        return 'Ultra Budget'
    if Price <= 15000:
        return 'Budget'
    if Price <= 28000:
        return 'Mid Range'
    if Price <= 40000:
        return 'Upper Range'
    if Price <= 65000:
        return 'Premium'
    else:
        return 'Flagship'
```

```python
data['Price Segment'] = data['Selling Price'].apply(Price_Segment)
```

```python
data.head()
```

```python
PS = data['Price Segment'].value_counts()
print('The price segment in India is as follows: ', PS)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
segment_counts = data['Price Segment'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
sns.barplot(x=segment_counts.index, y=segment_counts.values, palette='viridis')
# Titles and labels
plt.title("Number of Mobiles in Each Price Segment", fontsize=14)
plt.xlabel("Price Segment")
plt.ylabel("Number of Mobiles")

plt.xticks(rotation=30)

max_count = segment_counts.max()
plt.yticks(np.arange(0, max_count + 10, 20))  # Change 10 to 5 or 2 for more detail
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

### 3. Which brand caters to all different segments? (low range, mid range, premium)


```python
def segment_group(segment):
    if segment in ['Ultra Budget', 'Budget']:
        return 'Low Range'
    elif segment in ['Mid Range', 'Upper Middle']:
        return 'Mid Range'
    elif segment in ['Upper Middle', 'Premium']:
        return 'Upper Range'

data['Segment Group'] = data['Price Segment'].apply(segment_group)
```

```python
data.head()
```

```python
brand_segment_table = pd.crosstab(data['Brand'], data['Segment Group'])
```

```python
brand_segment_table
```

```python
brands_all_segment = brand_segment_table[
(brand_segment_table['Low Range']>0) &
(brand_segment_table['Mid Range']>0) &
(brand_segment_table['Upper Range']>0)
]
brands_all_segment
```

### 5. Provide more detailed insights based on the data provided.

```python
brand_stats = data.groupby('Brand').agg(
    Average_Selling_Price=('Selling Price', 'mean'),
    Average_Rating=('Rating', 'mean'),
    Number_of_Models=('Model', 'nunique') 
).sort_values(by='Average_Selling_Price', ascending=False) 

print("Average Selling Price and Rating per Brand (sorted by Average Selling Price, descending):")
print(brand_stats.to_markdown(numalign="left", stralign="left"))
```
