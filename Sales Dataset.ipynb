```python
L = ['Ship Mode', 'Segment', 'City', 'State', 'Region', 'Category', 'Sub-Category']
for i in L:
    
    print(df[i].value_counts())
    print()
```

```python
L1 = ['Ship Mode', 'Segment', 'City', 'State', 'Region', 'Category', 'Sub-Category']
for j in L1:
    
    print(df.groupby(j)['Sales'].agg(['sum', 'mean', 'count']))
    print()
```

```python
sales_by_category = df.groupby('Category')['Sales'].sum()
print(sales_by_category)
```

```python
sales_by_category.plot(kind = 'bar', figsize = (8,5))
plt.title('Total Sales by Category')
plt.ylabel('Sales')
plt.xlabel('Category')
plt.show()
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(5)
print(top_products)
```

```python
top_products.plot(kind='barh', figsize = (8,5), color='green')
plt.title('Top 5 products')
plt.xlabel('Sales')
plt.ylabel('Product Name')
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
```

```python
sales_over_time = df.groupby('Order Date')['Sales'].sum()
```

```python
sales_over_time.plot(figsize=(10,6))
plt.title('Sales Over Time')
plt.ylabel('Sales')
plt.xlabel('Order Date')
plt.show()
```
